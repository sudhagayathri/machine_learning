```python
word_punc = nltk.tokenize.WordPunctTokenizer()
word_punc.tokenize(comp_sentence)
```

```python
#stop words
stop_words = set(stopwords.words('english'))
```

```python
sent_filter = [x for x in comp_sentence if x not in stop_words]
```

```python
#to visualise while applying stop words
#above we are not able to see words properly afer removing stopwords, so we use word cloud to remove them
from wordcloud import WordCloud,STOPWORDS
```

```python
comp_sentence
```

```python
word_stop =nltk.corpus.stopwords.words('english')
```

```python
word_stop
```

```python
wcloud = WordCloud(max_words=50,stopwords=word_stop)
```

```python
wcloud.generate(comp_sentence)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.imshow(wcloud)
plt.show()
```

# Custimising stop words

```python
wcloud = WordCloud(max_words=50,stopwords=word_stop+["never", "also","along"], background_color="white")
```

```python
wcloud.generate(comp_sentence)
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.imshow(wcloud)
plt.show()
```
