```python
import pandas as pd #for dataframe
import numpy as np #mathematical calculation
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df =pd.read_csv("car_performance.csv")
df.head(20)
```

```python
df.info()
```

```python
df.describe()#along with mean, std, it also givs first quartile
```

```python
#distribution of country origin - count of cylinders of each type
plt.figure(figsize=(7,7)) 
sns.countplot(x="origin", data=df)
```

```python
#2 types of data - categorical, continous
#we do this graph of count for categorical value(where data can be grouped with limited types), 
#here weight , horsepower etc is continous, all rows are different
```

```python
#distribution of cylinders - count of cylinders of each type
plt.figure(figsize=(7,7))
sns.countplot(x="cylinders", data=df)
#here y axis is default count if not mentioned
```

```python
#distribution of model_year - count of model_year of each type
plt.figure(figsize=(7,7))
sns.countplot(x="model_year", data=df)
#here y axis is default count if not mentioned
```

```python
#ths is continous data - ideally it shud be not chosen, this will not give any information - we shud choose categorical data only
#distribution of weight - count of model_year of each type
plt.figure(figsize=(7,7))
sns.countplot(x="weight", data=df)
#here y axis is default count if not mentioned
```

```python
#subplot
```

```python
plt.subplots(2,2,figsize=(16,10)) # 2 rows, 2 columns
plt.subplot(2,2,1) #giving 1st postion
sns.countplot(x="origin", data=df)
plt.subplot(2,2,2)#giving 1st postion
sns.countplot(x="cylinders", data=df)
plt.subplot(2,2,3)
sns.countplot(x="model_year", data=df)
plt.subplot(2,2,4)
sns.countplot(x="mpg_level", data=df)
```

```python
#pie plot  -also to be plotted for categorical value
plt.figure(figsize=(5,5))
plt.pie(x=df['origin'].value_counts(),labels=df['origin'].value_counts().index, autopct="%0.2f%%", data=df)
```

```python
plt.subplots(2,2,figsize=(16,10))

plt.subplot(2,2,1)
plt.pie(x=df["origin"].value_counts(),labels=df["origin"].value_counts().index,autopct='%0.2f%%',data=df)
plt.title('Percentage of each origin in the data')

plt.subplot(2,2,2)
plt.title("Distribution of cylinder")
plt.pie(x=df["cylinders"].value_counts(),labels=df["cylinders"].value_counts().index,autopct='%0.2f%%',explode=[0,0,0,0.03,0.2],data=df)


plt.subplot(2,2,3)

plt.title("Distribution of model_year")
plt.pie(x=df["model_year"].value_counts(),labels=df["model_year"].value_counts().index,autopct='%0.2f%%',data=df)

plt.subplot(2,2,4)
plt.title("Distribution of mpg_level")
plt.pie(x=df["mpg_level"].value_counts(),labels=df["mpg_level"].value_counts().index,autopct='%0.2f%%',data=df)

plt.show()
```

```python
#observations
#usa alone contributes 63 per
#4 no cylinder contributes 50 per
#medium level mpg_cylinder contributes 53 per
```

```python
#Adding `hue="origin" tells seaborn you want to colour the data points for different origins differently.
plt.figure(figsize=(5,5))
sns.countplot(x="cylinders", data=df,hue="origin")
plt.title('Distribution of cylinders w.r.t country')
```

```python
#observations
#japan itself has 3 cylinder car
#europe alone have 5 cylinder car
#usa alone has 8 cylinder car
#all countries has 4 cylinder car
#all countries has 6 cylinder car, but dominated by usa
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x="mpg_level", data=df,hue="origin")
plt.title('Distribution of cylinders w.r.t country')
```

```python
#low mileage lvel is not there in japan, mileage is not important by japan
#where europe also has negligable low_mpg
#usa is dominating low and medium lvel
```

# Bar plot

```python
#till now we have seen count plot, it is against count, but here in bar plot- we can see max, min and evrything
#in count plot always y-axis is count, but in bar, we can pass x-axis and in y-axis we can see one more param
```

```python
df.head()
```

```python
plt.figure(figsize=(18,10))
sns.barplot(x="origin", y="displacement", data=df)
```

```python
#here the black mark on top of graph, indicates which is max point and min point
#like for usa, max is 260, min is 240
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x="origin", data=df)
```

```python
plt.figure(figsize=(18,10))
sns.barplot(x="origin", y="mpg", data=df, hue="cylinders")
```

```python
#line plot
```

```python
plt.subplots(2,3,figsize=(16,10))

plt.subplot(2,3,1)
sns.lineplot(x="model_year",y="mpg",hue="origin",data=df)
plt.xlabel('model year')


plt.subplot(2,3,2)
sns.lineplot(x="model_year",y="displacement",hue="origin",data=df)
plt.xlabel('model year')

plt.subplot(2,3,3)
sns.lineplot(x="model_year",y="horsepower",hue="origin",data=df)
plt.xlabel('model year')


plt.subplot(2,3,4)
sns.lineplot(x="model_year",y="weight",hue="origin",data=df)
plt.xlabel('model year')

plt.subplot(2,3,5)
sns.lineplot(x="model_year",y="acceleration",hue="origin",data=df)
plt.xlabel('model year')

plt.subplot(2,3,4)
sns.lineplot(x="model_year",y="weight",hue="origin",data=df)
plt.xlabel('model year')

plt.show()
```

```python
#with increasing year, usa is reducing weight
```

# Histogram - plays major role in ml, 

```python
#its useful to see whether data is following normal distribution/not
#whether data is symmetrical/not 
#we can use this on NUMERICAL DATA to see how the data is
```

```python
plt.figure(figsize=(8,6))
sns.histplot(x="mpg",data=df)
```

```python
plt.figure(figsize=(8,6))
sns.histplot(x="mpg",data=df,kde=True)
```

```python
# the above shows its not SYMMETRICAL - u cant equally divide in 2 parts
#the above is right skewd - tail is towards right
```
