```python
corr=df.corr() #by default pearson approach
corr
```

```python
#pearson is parametric approach - where we assume, distribution is linear, dispersion is less
```

```python
corr1=df.corr(method='spearman')
corr1
```

```python
#if it is monotonic then folow kendall
corr2=df.corr(method='kendall')
corr2
```
