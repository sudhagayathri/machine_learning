```python
def univariate_analysis(col_list):
    fig, axs = plt.subplots(1,2,figsize=(10,3))
    for index, column_name in enumerate(col_list):
        df_temp = pd.DataFrame(sorted(list(df[column_name].unique())),columns=[column_name])
        df_temp['Total_Count'] = list(df.groupby(column_name).Accept.count())
        df_temp['Total_Count_per'] = list(df.groupby(column_name).Accept.count())
        df_temp['Accepted'] = list(df[df.Accept==1].groupby(column_name).Accept.count())
        df_temp['Rejected'] = list(df[df.Accept==0].groupby(column_name).Accept.count())
        df_temp['%Accepted'] = round(df_temp['Accepted']/df_temp['Total_Count']*100,3)
        df_temp['%Rejected'] = round(df_temp['Rejected']/df_temp['Total_Count']*100,3)
        df_temp = df_temp.sort_values(by='Total_Count', ascending=False)

        axs[index].barh(df_temp[column_name],df_temp['Total_Count'])
        axs[index].barh(df_temp[column_name],df_temp['Accepted'])
            # Add the percentage labels on top of the bars
        for i, v in enumerate(df_temp['%Accepted']):
            if v >0:
                axs[index].text(v + 1, i - 0.1, str(round(v)) + '%')

        axs[index].legend(labels=['Total','Accepted'])
        axs[index].set_xlabel('Counts')
        axs[index].set_title(column_name+' wise accepted coupons')
    plt.show()
```

```python
def plot_auc(model_name, roc_param):
    plt.figure(1)
    lw = 2
    plt.plot(fpr_dt, tpr_dt, color='green',
             lw=lw, label=model_name+'(AUC = %0.2f)' % roc_param)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Area Under Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
import itertools
def plot_confusion_matrix(model, normalize=False): # This function prints and plots the confusion matrix.
    cm = confusion_matrix(y_test, model, labels=[0, 1])
    classes=[" Not Accepted", "Accepted"]
    cmap = plt.cm.Blues
    title = "Confusion Matrix"
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        cm = np.around(cm, decimals=3)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```
