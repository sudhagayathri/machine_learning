```python
# Society
df['society'].value_counts()
```

```python
# cleaning society column
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★','',
                                                        str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df['price'].value_counts()
```

```python
df[df['price']=='Price on Request'].shape
```

```python
df = df[df['price']!='Price on Request']
```

```python
df.shape
```

```python
def treat_price(x):
    if type(x)==float:
        return x
    else:
        if x[1] =='Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.head()
```

```python
df['Price_per_sqft'].value_counts()
```

```python
df['Price_per_sqft'] = df['Price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
df.head()
```
